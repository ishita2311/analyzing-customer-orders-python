# Analyzing Customer Orders Using Python

Course-End Project

## Objective
Analyze customer orders to extract business insights using Python data structures.

```

customers = ["Alice", "Bob", "Charlie", "Diana", "Eve"]

orders = [
    ("Alice", "Laptop", 800, "Electronics"),
    ("Alice", "Headphones", 50, "Electronics"),
    ("Bob", "T-shirt", 25, "Clothing"),
    ("Bob", "Jeans", 45, "Clothing"),
    ("Charlie", "Microwave", 120, "Home Essentials"),
    ("Charlie", "Blender", 60, "Home Essentials"),
    ("Diana", "Smartphone", 600, "Electronics"),
    ("Diana", "Dress", 70, "Clothing"),
    ("Eve", "Curtains", 40, "Home Essentials"),
]

customer_orders = {}
for name, product, price, category in orders:
    customer_orders.setdefault(name, []).append((product, price, category))
```

```

customer_spending = {}
customer_type = {}

for customer, items in customer_orders.items():
    total = sum(price for _, price, _ in items)
    customer_spending[customer] = total

    if total > 100:
        customer_type[customer] = "High-Value Buyer"
    elif 50 <= total <= 100:
        customer_type[customer] = "Moderate Buyer"
    else:
        customer_type[customer] = "Low-Value Buyer"

customer_spending, customer_type
```

```

category_revenue = {}
for _, _, price, category in orders:
    category_revenue[category] = category_revenue.get(category, 0) + price

category_revenue
```
